```python
import os
import sys
import pandas as pd

sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from scripts.data_preprocessing import *
```

## Sentiments Cleaning

```python
sentiment_train = pd.read_csv("../data/raw/train_data.csv")
sentiment_test = pd.read_csv("../data/raw/test_data.csv")
```

```python
sentiment_test["sentence"] = sentiment_test["sentence"].fillna("").astype(str)
sentiment_train["sentence"] = sentiment_train["sentence"].fillna("").astype(str)
```

```python
cleaned_sentiment_train = preprocess_sentiment(sentiment_train)
cleaned_sentiment_test = preprocess_sentiment(sentiment_test)
```

```python
cleaned_sentiment_test.head()
```

## G0emotions Cleaning

```python
goemotions1 = pd.read_csv("../data/raw/goemotions_1.csv")
goemotions2 = pd.read_csv("../data/raw/goemotions_2.csv")
goemotions3 = pd.read_csv("../data/raw/goemotions_3.csv")
```

```python
goemotions1["text"] = goemotions1["text"].fillna("").astype(str)
goemotions2["text"] = goemotions2["text"].fillna("").astype(str)
goemotions3["text"] = goemotions3["text"].fillna("").astype(str)
```

```python
cleaned_goemotions1 = preprocess_goemotions(goemotions1)
cleaned_goemotions2 = preprocess_goemotions(goemotions2)
cleaned_goemotions3 = preprocess_goemotions(goemotions3)
```

## Artifacts

```python
cleaned_sentiment_train.to_csv("../data/cleaned/cleaned_sentiment_train.csv", index=False)
cleaned_sentiment_test.to_csv("../data/cleaned/cleaned_sentiment_test.csv", index=False)
cleaned_goemotions1.to_csv("../data/cleaned/cleaned_goemotions1.csv", index=False)
cleaned_goemotions2.to_csv("../data/cleaned/cleaned_goemotions2.csv", index=False)
cleaned_goemotions3.to_csv("../data/cleaned/cleaned_goemotions3.csv", index=False)
```

```python
appended_goemotions = pd.concat([cleaned_goemotions1, cleaned_goemotions2, cleaned_goemotions3], axis=0)
appended_goemotions.to_csv("../data/appended/goemotions.csv", index=False)
```
